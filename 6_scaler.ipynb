```python
import pandas as pd 

s1 = pd.Series([10,20,30,40,50])
s2 = pd.Series([1,2,3,4,5])
print(s1)
print(s2)
```

```python
# addition
print("addition of 2 series \n",s1.add(s2))
```

```python
#subtraction 
print("subtraction of 2 series\n",s1.sub(s2))
```

```python
#multiplication
print("multiplication of 2 series\n",s1.mul(s2))
```

```python

#division
print("Division of 2 series \n",s1.div(s2))
```

```python
#power
print("Power of Series by 2 \n",s1.pow(3))
```

```python
#mean
print("Mean of Series",s1.mean())
#median
print("Median of Series ",s1.median())
#sum
print("Sum of series ",s1.sum())
```
